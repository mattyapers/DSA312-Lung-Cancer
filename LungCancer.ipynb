# General Demographic Trends

## Lung Cancer Prevelance

```python
# Sample of 9000, Lung Cancer Prevelance
lung_cancer = df['LUNG_CANCER'].value_counts()
lung_cancer

# Plot
plt.figure(figsize=(8,6))
plt.pie(lung_cancer,
        labels=['Yes', 'No'],
        autopct='%1.1f%%',
        startangle=90,
        explode=[0,0.1],
        shadow=True)

plt.title('Do you have Lung Cancer?', fontsize = 14)
plt.show()
```

## Lung Cancer Prevelance by Gender

```python
gender_lung_cancer_y = df[df['LUNG_CANCER'] == 1]['GENDER'].value_counts()

# Visualisation
plt.figure(figsize=(8,8))

plt.pie(gender_lung_cancer_y,
        labels=['Male', 'Female'],
        autopct='%1.1f%%',
        startangle=90,
        explode=[0,0.1],
        shadow=True)

plt.title('Proportion of Lung Cancer Patients by Gender')
```

## Lung Cancer Prevelance by Gender and by Age

```python
# Male Lung Cancer Patients
male_lung_cancer_y = df[(df['GENDER'] == 1) & (df['LUNG_CANCER'] == 1)]

# Female Lung Cancer Patients
female_lung_cancer_y = df[(df['GENDER'] == 0) & (df['LUNG_CANCER'] == 1)]
```

```python
# Male Lung Cancer Patients
plt.figure(figsize=(10,8))

# Calculate the youngest and oldest male ages and their frequencies
min_male_age = male_lung_cancer_y['AGE'].min()
min_male_age_freq = (male_lung_cancer_y['AGE'] == min_male_age).sum()

max_male_age = male_lung_cancer_y['AGE'].max()
max_male_age_freq = (male_lung_cancer_y['AGE'] == max_male_age).sum()

# Create histogram with KDE for male lung cancer patients
sns.histplot(
    x='AGE',
    data=male_lung_cancer_y,
    bins=20,
    kde=True,
    alpha=0.5    # Set transparency for the histogram
)

# Set plot title and labels
plt.title('Age Distribution of Male Lung Cancer Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Annotate youngest age for males
plt.annotate(
    f'Youngest Male: {min_male_age} Years Old',
    xy=(min_male_age, min_male_age_freq),
    xytext=(min_male_age - 10, min_male_age_freq + 15),
    arrowprops=dict(color='black', arrowstyle='->'),
    fontsize=10,  # Adjust font size for better visibility
    color='black'  # Set text color
)

# Annotate oldest age for males
plt.annotate(
    f'Oldest Male: {max_male_age} Years Old',
    xy=(max_male_age, max_male_age_freq),
    xytext=(max_male_age + 2, max_male_age_freq + 5),
    arrowprops=dict(color='black', arrowstyle='->'),
    fontsize=10,
    color='black'
)

# Show the plot
plt.tight_layout()  # Adjust layout for better spacing
plt.show()  # Render the plot
```

```python
# Male Lung Cancer Patients
plt.figure(figsize=(10,10))

## Youngest Female and Oldest Female
min_female_age = female_lung_cancer_y['AGE'].min()
min_female_age_freq = len(female_lung_cancer_y[female_lung_cancer_y['AGE'] == min_female_age])

max_female_age = female_lung_cancer_y['AGE'].max()
max_female_age_freq = len(female_lung_cancer_y[female_lung_cancer_y['AGE'] == max_female_age])

sns.histplot(
    x='AGE',
    data=female_lung_cancer_y,
    bins=20,
    kde=True,
    alpha = 0.5
)

plt.title('Age Distribution of Female Lung Caner Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')


## Annotate youngest age for Male
plt.annotate(
    f'Youngest Female : {min_female_age} Years Old',
    xy=(min_female_age,min_female_age_freq),
    xytext=(min_female_age - 10, min_female_age_freq + 10),
    arrowprops=dict(color='black',
                    arrowstyle='->')
)
## Annotate oldest age for Male
plt.annotate(
    f'Oldest Female : {max_female_age} Years Old',
    xy=(max_female_age, max_female_age_freq),
    xytext=(max_female_age + 2, max_female_age_freq + 5),
    arrowprops=dict(color='black',
                    arrowstyle='->')
)
```

## Relationship between Gender, Peer Pressure and Lung Cancer

```python
gender_peer_age_lung = df[['GENDER','PEER_PRESSURE', 'AGE', 'LUNG_CANCER']]
gender_peer_age_lung['GENDER'] = gender_peer_age_lung['GENDER'].replace({0 : 'Female', 1 : 'Male'})
```

```python
# Group by GENDER and aggregate PEER_PRESSURE and LUNG_CANCER with sum and count
grouped_data = gender_peer_age_lung.groupby('GENDER')[['PEER_PRESSURE', 'LUNG_CANCER']].agg(['sum', 'count']).reset_index()
grouped_data = grouped_data.drop(columns=[('PEER_PRESSURE', 'count')])

# Calculate the proportions
peer_pressure_proportion = grouped_data[('PEER_PRESSURE', 'sum')] / grouped_data[('LUNG_CANCER', 'count')]
lung_cancer_proportion = grouped_data[('LUNG_CANCER', 'sum')] / grouped_data[('LUNG_CANCER', 'count')]
susceptible_proportion = grouped_data[('PEER_PRESSURE', 'sum')] / grouped_data[('LUNG_CANCER', 'sum')]

ax = grouped_data.plot(
    kind='bar',
    x='GENDER',
    figsize=(10, 6),
    color=['skyblue', 'lightgreen',  'orange'],
    title='Peer Pressure and Lung Cancer by Gender'
)

ax.set_xlabel('Gender')
ax.set_ylabel('Counts')
ax.set_title('Peer Pressure and Lung Cancer Cases by Gender')
ax.legend(['Peer Pressure (Yes)','Lung Cancer (Yes)','Total Gender Population'])
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```
