## Relationship between Gender, Peer Pressure and Lung Cancer

```python
gender_peer_age_lung = df[['GENDER','PEER_PRESSURE', 'AGE', 'LUNG_CANCER']]
gender_peer_age_lung['GENDER'] = gender_peer_age_lung['GENDER'].replace({0 : 'Female', 1 : 'Male'})
```

```python
# Group by GENDER and aggregate PEER_PRESSURE and LUNG_CANCER with sum and count
grouped_data = gender_peer_age_lung.groupby('GENDER')[['PEER_PRESSURE', 'LUNG_CANCER']].agg(['sum', 'count']).reset_index()
grouped_data = grouped_data.drop(columns=[('PEER_PRESSURE', 'count')])

# Calculate the proportions
peer_pressure_proportion = grouped_data[('PEER_PRESSURE', 'sum')] / grouped_data[('LUNG_CANCER', 'count')]
lung_cancer_proportion = grouped_data[('LUNG_CANCER', 'sum')] / grouped_data[('LUNG_CANCER', 'count')]
susceptible_proportion = grouped_data[('PEER_PRESSURE', 'sum')] / grouped_data[('LUNG_CANCER', 'sum')]

ax = grouped_data.plot(
    kind='bar',
    x='GENDER',
    figsize=(10, 6),
    color=['skyblue', 'lightgreen',  'orange'],
    title='Peer Pressure and Lung Cancer by Gender'
)

ax.set_xlabel('Gender')
ax.set_ylabel('Counts')
ax.set_title('Peer Pressure and Lung Cancer Cases by Gender')
ax.legend(['Peer Pressure (Yes)','Lung Cancer (Yes)','Total Gender Population'])
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

# Symptom Analysis

### Common symptoms from Lung Cancer Patients

```python
lung_cancer_y = df[df['LUNG_CANCER']==1]
lung_cancer_y.info()
lung_cancer_symptoms = lung_cancer_y[['YELLOW_FINGERS', 'FATIGUE ', 'WHEEZING', 
                                      'COUGHING', 'SHORTNESS OF BREATH', 
                                      'SWALLOWING DIFFICULTY', 'CHEST PAIN']]
# lung_cancer_symptoms.replace({0 : 'No', 1 : 'Yes'}, inplace= True)
```

```python
# Melt the data to have a suitable format for plotting
lung_cancer_symptoms_melted = lung_cancer_symptoms.melt(var_name='Symptom', value_name='Presence')

# Count the number of patients with each symptom
symptom_counts = lung_cancer_symptoms_melted.groupby('Symptom')['Presence'].sum()

# Sort symptoms by count to determine the order
sorted_symptom_counts = symptom_counts.sort_values(ascending=False)

# Determine the most common symptom
most_common_symptom = sorted_symptom_counts.idxmax()
most_common_count = sorted_symptom_counts.max()
total_patients = len(lung_cancer_y)
most_common_percentage = (most_common_count / total_patients) * 100

# Plot the countplot
plt.figure(figsize=(10, 10))
ax = sns.countplot(x='Symptom', hue='Presence', data=lung_cancer_symptoms_melted)

# Change legend from 0.0, 1.0 to No, Yes
ax.legend(title='Presence', labels=['No', 'Yes'])

# Annotate the count on top of each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='baseline', 
                fontsize=10, color='black', 
                xytext=(0, 3),  # 3 points vertical offset
                textcoords='offset points')
    

plt.title('Symptom Distribution Among Lung Cancer Patients')
plt.ylabel('Count')
plt.xlabel('Symptoms')
plt.xticks(rotation=45)
plt.tight_layout()

# Print out the most common symptom and its percentage
print(f'Most common symptom: {most_common_symptom}, '
      f'Percentage of lung cancer patients: {most_common_percentage:.1f}%')

# Print the order of common symptoms
for symptom, count in sorted_symptom_counts.items():
    percentage = (count / total_patients) * 100
    print(f'Symptom: {symptom}, Count: {count}, Percentage: {percentage:.1f}%')

plt.show()
```
