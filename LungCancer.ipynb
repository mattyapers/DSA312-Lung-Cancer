# General Demographic Trends

## Lung Cancer Prevelance

```python
# Sample of 9000, Lung Cancer Prevelance
lung_cancer = df['LUNG_CANCER'].value_counts()
lung_cancer

# Plot
plt.figure(figsize=(8,6))
plt.pie(lung_cancer,
        labels=['Yes', 'No'],
        autopct='%1.1f%%',
        startangle=90,
        explode=[0,0.1],
        shadow=True)

plt.title('Do you have Lung Cancer?', fontsize = 14)
plt.show()
```

## Lung Cancer Prevelance by Gender

```python
gender_lung_cancer_y = df[df['LUNG_CANCER'] == 1]['GENDER'].value_counts()

# Visualisation
plt.figure(figsize=(8,8))

plt.pie(gender_lung_cancer_y,
        labels=['Male', 'Female'],
        autopct='%1.1f%%',
        startangle=90,
        explode=[0,0.1],
        shadow=True)

plt.title('Proportion of Lung Cancer Patients by Gender')
```

## Proportion within Gender that have Lung Cancer

```python
# Male Respondents
male = df[df['GENDER'] == 1]['LUNG_CANCER'].value_counts()

# Female Respondents
female = df[df['GENDER'] == 0]['LUNG_CANCER'].value_counts()
```

```python
# Visualization for Male
plt.figure(figsize=(10,8))

plt.pie(male,
        labels=['Yes', 'No'],
        autopct='%1.1f%%',
        startangle=90,
        explode=[0,0.1],
        shadow=True)

plt.title('Lung Cancer in Male')
plt.show()
```

```python
# Visualization for Female
plt.figure(figsize=(10,8))

plt.pie(female,
        labels=['Yes', 'No'],
        autopct='%1.1f%%',
        startangle=90,
        explode=[0,0.1],
        shadow=True)

plt.title('Lung Cancer in Female')
plt.show()
```

## Lung Cancer Prevelance by Gender and by Age

```python
# Male Lung Cancer Patients
male_lung_cancer_y = df[(df['GENDER'] == 1) & (df['LUNG_CANCER'] == 1)]

# Female Lung Cancer Patients
female_lung_cancer_y = df[(df['GENDER'] == 0) & (df['LUNG_CANCER'] == 1)]
```

```python
# Male Lung Cancer Patients
plt.figure(figsize=(10,10))

## Youngest Male and Oldest Male
min_male_age = male_lung_cancer_y['AGE'].min()
min_male_age_freq = len(male_lung_cancer_y[male_lung_cancer_y['AGE'] == min_male_age])

max_male_age = male_lung_cancer_y['AGE'].max()
max_male_age_freq = len(male_lung_cancer_y[male_lung_cancer_y['AGE'] == max_male_age])

sns.histplot(
    x='AGE',
    data=male_lung_cancer_y,
    bins=20,
    kde=True
)

plt.title('Age Distribution of Male Lung Caner Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')

# # Annotate highest sales point for Product A
# max_a = df['Product_A_Sales'].max()
# max_a_month = df[df['Product_A_Sales'] == max_a]['Month'].values[0]
# plt.annotate(f'Max: {max_a}',
#              xy=(max_a_month, max_a),
#              xytext=(max_a_month, max_a+15),
#              arrowprops=dict(color='black',
#                              arrowstyle='->')) 

## Annotate youngest age for Male
plt.annotate(
    f'Youngest Male : {min_male_age} Years Old',
    xy=(min_male_age,min_male_age_freq),
    xytext=(min_male_age - 10, min_male_age_freq + 10),
    arrowprops=dict(color='black',
                    arrowstyle='->')
)
## Annotate oldest age for Male
plt.annotate(
    f'Oldest Male : {max_male_age} Years Old',
    xy=(max_male_age, max_male_age_freq),
    xytext=(max_male_age + 2, max_male_age_freq + 5),
    arrowprops=dict(color='black',
                    arrowstyle='->')
)
```
